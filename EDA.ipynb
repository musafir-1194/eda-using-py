### cleaning the data

```python
data.isnull().sum()
```

```python
# removing columns which are not required
remove_cols = ['name', 'nationality', 'city', 'gender','ethnic.group', 'latitude', 'longitude', 'portfolio.rating', 'coverletter.rating', 'refletter.rating']
student = data.drop(remove_cols, axis=1)
```

```python
print(student.head())
print(student.nunique)
```

#### Relationship analysis

```python
student.isnull().sum()
```

```python
corelation = student.corr()
```

```python
sns.heatmap(corelation, xticklabels=corelation.columns, yticklabels=corelation.columns,
           annot=True)
```

```python
sns.pairplot(student)
```

```python
sns.relplot(x='math.grade', y='english.grade', hue='age', data=student)
```
